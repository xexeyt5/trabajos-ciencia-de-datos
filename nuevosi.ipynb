```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

import warnings
warnings.filterwarnings("ignore")
```

```python
cwd = os.getcwd()

os.chdir('/home/xexeyt/Desktop/trabajos/')

population = pd.read_csv('employee_salary.csv', header=0, sep=',')
population
```

```python
population_base_salary = population['Base Salary']

sns.distplot(population_base_salary)
plt.xlabel('population_base_salary')
plt.ylabel('population_base_salary')
```

```python
mean = []

def calc_sample_mean(data, samble_size, no_sample_menas):
    for i in range(no_sample_menas):
        sample = data.sample(n=samble_size)
        sample_mean = sample.mean()
        mean.append(sample_mean)
    return(mean)
```

```python
mean_2 = calc_sample_mean(data=population_base_salary, samble_size=2, no_sample_menas=500)

sns.distplot(mean_2, color='b')
plt.xlabel('sample base salary (sample size =2)')
plt.ylabel('probability density')
```

```python
mean_3 = calc_sample_mean(data=population_base_salary, samble_size=5, no_sample_menas=500)

sns.distplot(mean_3, color='r')
plt.xlabel('sample base salary (sample size =5)')
plt.ylabel('probability density')
```
